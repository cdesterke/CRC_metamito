```r
data<-small[["RNA"]]$counts
head(data)[,1:5]
```

```r
data(mitol2)
mito<-mitogenescore(data,symbol=mitol2,method="ssgsea")
head(mito)[,1:5]
```

```r
mito<-as.matrix(mito)
all(colnames(mito)==colnames(small))
```

```r
assay <- CreateAssayObject(data = mito)
small[["mitochondria"]] <- assay


small[["mitochondria"]]
```

```r
options(repr.plot.width = 20, repr.plot.height = 8)
DefaultAssay(small)<-"mitochondria"
FeaturePlot(small, features = c("ABC-transporters"),min.cutoff = "q9",cols=c("#CCFFFF","black"),
reduction="umap",pt.size=0.5,split.by="orig.ident")
```

```r
vector<-row.names(small$mitochondria)
vector
```

```r
library(viridis)
library(RColorBrewer)
library(ggplot2)
```

```r
colgex=c("grey90",brewer.pal(7,"Reds"))
plotTheme=theme_classic(base_size=16)
```

```r
options(repr.plot.width = 14, repr.plot.height = 12)
DotPlot(small,group.by="orig.ident",features=vector)+
 scale_color_gradientn(colors = colgex) +
  theme(axis.text.x = element_text(angle = -45, hjust = 0))+coord_flip()
```
